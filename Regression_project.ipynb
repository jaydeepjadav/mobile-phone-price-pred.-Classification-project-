````python
#copying data to new dataframe variable
new_df = df.copy()
````

Extracting date and time data in more useable form

````python
import datetime
````

````python
#combining date and time column into one
new_df['date_time'] = new_df['travel_date'] + " " + new_df['travel_time']
new_df['date_time'] = pd.to_datetime(new_df['date_time'])
````

````python
#dropping existing date column and time column
new_df.drop(['travel_date','travel_time'],inplace=True,axis=1)
````

````python
#extracting required date and time columns information

new_df['month'] = new_df['date_time'].dt.month
new_df['year'] = new_df['date_time'].dt.year
new_df['hour'] = new_df['date_time'].dt.hour
new_df['day_of_year'] = new_df['date_time'].dt.day_of_year
new_df['day_of_month'] = new_df['date_time'].dt.day
new_df['day_of_week'] = new_df['date_time'].dt.dayofweek
new_df['quarter_of_year'] = new_df['date_time'].dt.quarter
new_df['weekend_travels'] = new_df['date_time'].apply( lambda x: 1 if x in [5,6] else 0)
````

````python
new_df.nunique()
````

````python
def time_of_day(i):
  if i<7 :
    return 'early morning'
  elif i>=7 and i<12:
    return 'morning'
  elif i>=12 and i<16:
    return  'afternoon'
  elif i>=16 and i<19:
    return  'evening'
  else:
    return  'night'
````

````python
# this column will show which time period of day
new_df['travelling_period'] = new_df.hour.apply(time_of_day)
````

````python
temp_var = dict(new_df.travelling_period.value_counts())
new_df['hourly_travellers'] = np.log1p(new_df.travelling_period.map(temp_var))
````

````python
# creating new coulmn for daily travellers
temp_var2 = dict(new_df.day_of_year.value_counts())
new_df['daily_travellers'] = np.log1p(new_df.day_of_year.map(temp_var2))
````

````python
#Creating a function to calculate frequancy of buses
def calculate_next_and_previous_timings(tmp_df):

  tmp_df.sort_values(['travel_from', 'date_time'], inplace= True)
  tmp_df['delay_btw_initial_to_next_and_previous_bus'] = (tmp_df.groupby(['travel_from']).date_time.shift(-1) - tmp_df.groupby(['travel_from']).date_time.shift(1)).dt.total_seconds()/3600
  tmp_df['delay_btw_1bus_and_next_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(-1)).dt.total_seconds()/3600
  tmp_df['delay_btw_1bus_and_previous_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(1)).dt.total_seconds()/3600
  tmp_df['delay_btw_2bus_and_next_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(-2)).dt.total_seconds()/3600
  tmp_df['delay_btw_2bus_and_previous_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(2)).dt.total_seconds()/3600
  tmp_df['delay_btw_3bus_and_next_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(-3)).dt.total_seconds()/3600
  tmp_df['delay_btw_3bus_and_previous_bus'] = (tmp_df.date_time - tmp_df.groupby(['travel_from']).date_time.shift(3)).dt.total_seconds()/3600
  new_col = ['delay_btw_initial_to_next_and_previous_bus', 'delay_btw_1bus_and_next_bus', 'delay_btw_1bus_and_previous_bus', 'delay_btw_2bus_and_next_bus', 'delay_btw_2bus_and_previous_bus', 'delay_btw_3bus_and_next_bus','delay_btw_3bus_and_previous_bus']
  tmp_df[new_col] = tmp_df.groupby(['travel_from'])[new_col].fillna(method = 'ffill')
  tmp_df[new_col] = tmp_df.groupby(['travel_from'])[new_col].fillna(method = 'backfill')

  return tmp_df
````

````python
new_df = calculate_next_and_previous_timings(new_df)
````
